```python
# Dependencies
import pandas as pd
# load CSV
file_to_load="Resources/purchase_data.csv"
```

```python
# read with pandas
file_df=pd.read_csv(file_to_load)
file_df.head()
```

```python
# Total Number of Players
Total_players=len(file_df["SN"].unique())
Summary_Total_players=pd.DataFrame({"Total Players": [Total_players]})
Summary_Total_players
```

Purchasing Analysis (Total)

* Number of Unique Items
* Average Purchase Price
* Total Number of Purchases
* Total Revenue

```python
Unique_items=len(file_df["Item Name"].unique())
average_price=round(file_df["Price"].mean())
number_of_purchases=len(file_df["Purchase ID"].unique())
total_price=file_df["Price"].sum()
summary_purchasing_analysis=pd.DataFrame({"Number of Unique Items":[Unique_items],
                            "Average Price" :average_price,
                            "Number of Purchases": number_of_purchases,
                            "Total Revenue":total_price})
summary_purchasing_analysis
```

Gender Demographics

* Percentage and Count of Male Players
* Percentage and Count of Female Players
* Percentage and Count of Other / Non-Disclosed

```python
s=file_df.Gender
count=s.value_counts()
percent100=s.value_counts(normalize=True).mul(100).round(1).astype(str)+"%"
pd.DataFrame({'Total Count': count,
              'Percentage of Players': percent100})
```

Purchasing Analysis (Gender)
* The below each broken by gender
   * Purchase Count
   * Average Purchase Price
   * Total Purchase Value
   * Average Purchase Total per Person by Gender

```python

groupby_gender=file_df.groupby(['Gender'])
groupby_gender.count().head()
groupby_gender["Price"].sum()
groupby_gender["Purchase ID"].count()
groupby_gender["Price"].mean()
pd.DataFrame({'Purchase Count':groupby_gender["Purchase ID"].count(),
              'Average Purchase Price':groupby_gender["Price"].mean(),
              'Total Purchase Value': groupby_gender["Price"].sum()})
```

Age Demographics
 * The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)
 * Purchase Count
 * Average Purchase Price
 * Total Purchase Value
 * Average Purchase Total per Person by Age Group

```python
bins=[0, 9, 14, 19, 24, 29, 34, 39, 100]
group =["<10", "10-14", "15-19","20-24", "25-29", "30-34", "35-39", "40+"]
# groupby_age=file_df.groupby(['Age'] ,group )
file_df["Age Ranges"]=pd.cut(file_df["Age"], bins, labels=group)

file_df.head(3)
```

Purchasing Analysis (Gender)

* The below each broken by gender

  * Purchase Count
  * Average Purchase Price
  * Total Purchase Value
  * Average Purchase Total per Person by Gender

```python
groupby_age=file_df.groupby(['Age Ranges'])
pd.DataFrame({'Total Count':groupby_age["Age"].count(),
              'Percentage of Players': groupby_age["Age"].sum()/groupby_age["Age"].count()})
              
```

Top Spenders

* Identify the the top 5 spenders in the game by total purchase value, then list (in a table):

  * SN
  * Purchase Count
  * Average Purchase Price
  * Total Purchase Value

```python
groupby_age=file_df.groupby(['Age Ranges'])

pd.DataFrame({'Purchase Count':groupby_age["Purchase ID"].count(),
              'Percentage of Players': groupby_age["Price"].mean(),
              'Total Purchase Value': groupby_age["Purchase ID"].count()* groupby_age["Price"].mean(),
              'Ave Total Purchase per Person':groupby_age["Purchase ID"].count()* groupby_age["Price"].mean()})




```

Most Popular Items

* Identify the 5 most popular items by purchase count, then list (in a table):

  * Item ID
  * Item Name
  * Purchase Count
  * Item Price
  * Total Purchase Value

```python
Item_data=file_df.loc[:, ['Item ID', 'Item Name', 'Price']]
groupby_item = Item_data.groupby(['Item ID', 'Item Name'])
groupby_item.head()
total_item_purchase=groupby_item.sum()['Price']
total_item_purchase
average_item_purchase=groupby_item.mean()["Price"]
item_count=groupby_item.count()["Price"]

Item_dataframe=pd.DataFrame({'Total Puchase Value':total_item_purchase,
              'Average Item Puchase': average_item_purchase,
              'Item Count': item_count,
              })

Sorted_items=Item_dataframe.sort_values('Item Count', ascending=False)
Sorted_items.head(5)
```

Most Profitable Items

* Identify the 5 most profitable items by total purchase value, then list (in a table):

  * Item ID
  * Item Name
  * Purchase Count
  * Item Price
  * Total Purchase Value

```python
Item_dataframe=pd.DataFrame({'Total Puchase Value':total_item_purchase,
              'Average Item Puchase': average_item_purchase,
              'Item Count': item_count,
              })

Sorted_items_profitable=Item_dataframe.sort_values('Total Puchase Value', ascending=False)
Sorted_items_profitable.head(5)
```
